# Анализ законономерностей, определяющих успешность игры

**Описание данных:**

## Шаг 2. Подготовка данных

- **Замена названий столбцов**

Преобразуем названия столбцов, приведя значения к нижнему регистру. Аббривеатуры регионов продаж оставим неизменными, так как большие буквы в них имеют значение.

- **Обработка пропусков**

1. Для многих записей пропущен год издания. Посмотрим повнимательнее на список этих игр.

```python
no_year = games_df[games_df['year_of_release'].isna()].copy()
```

```python
no_year['name'].tolist()
```

Видно, что большая часть этих игр достаточно известны (например Space Invaders 1973 года), соотвтетственно известен и год их издания, однако по какой-то ошибке в наших данных эти даты пропали. Можно было бы восстановить их вручную, но это не так интересно, поэтому воспользуется API сайта RAWG, который предоставляет информацию более чем о 300 тысяч игр. 

Документация: https://api.rawg.io/docs/

Используем обертку библиотеки RAWGpy

Библиотека: https://pypi.org/project/rawgpy/

Документация: https://rawgpy.readthedocs.io/en/latest/index.html

```python
pip install rawgpy
```

```python
import rawgpy
```

Напишем функцию, получающую дату издания игры из API. 

```python
def get_year(row):
    rawg = rawgpy.RAWG("User-Agent")
    game_name = row['name']
    results = rawg.search(game_name)  # по умолчанию возвращает 5 наиболее релевантных результатов 
    game = results[0] # поэтому берем первый 
    try: 
        return game.released
    except: 
        return np.nan
```

Применим функцию к копии нашего датафрейма

```python
no_year['year']=no_year.apply(get_year, axis=1)
```

Данные получены в виде DD-MM-YYYY, поэтому переведем им в datetime и вытянем из него год издания.

```python
no_year['year']= pd.to_datetime(no_year['year'])
```

```python
no_year['year_of_release'] = pd.DatetimeIndex(no_year.year).year
```

Заменим пропущенные данные в столбце 'year_of_release'

```python
games_df.loc[games_df['year_of_release'].isna(),['year_of_release']] = no_year['year_of_release']
```

Некоторое количество игр так и не получило год издания. Таких оказалось всего 9, и их удаление не значительно повлияет на качество данных, так что их можно безболезненно отбросить. 

```python
games_df.dropna(subset=['year_of_release'], inplace=True)
```

2. Так же уберем неизвестные игры для genesis из столбца 'name'

```python
games_df.dropna(subset=['name'], inplace=True)
```

3. Путем сравнения было определено, что в столбце critic_score дан так называемый metascore, полученных из данных сайта Metacritic. Некоторую часть значений можно так же востановить с помощью API RAWG. (адекватного варпера на питоне для метакритика я не нашел, хехе)

Определим функцию для получения metascore с сайта RAWG.

```python
def get_critic_score(row):
    rawg = rawgpy.RAWG("User-Agent")
    game_name = row['name']
    results = rawg.search(game_name)
    try: 
        game = results[0]
        return game.metacriric
    except: 
        return np.nan
```

В итоге оказалось что новых значений в столбец с critic_score это не принесло. Но мы пытались.  

Пропуски с оценками критиков решено не заполнять. Они практически не участвуют в оценке и для необходимого анализа достаточно имеющихся данных. К тому же вероятнее всего многие из этих игр просто достаточно редки, чтобы иметь оценку на метакритике.

4. Столбец с user_score можно было бы заполнить подобным образом (рейтинг RAWG существует практически для всех игр), однако в RAWG используется 5 бальная система оценки, что может привести к расхождениям с имеющимися данными. (Можно подгрузить данные для всех игр в отдельный столбец RAWG_user_score, но это будет ОЧЕНЬ долго).

Пропуски проигнорируем, и заметим, что в списке значений этого столбца есть численное значение TBD. Для корректной обработки его стоит убрать. Что оно может означать? Заглянем в FAQ метакритика  

https://www.metacritic.com/faq#item13  

ТBD - аббривеатура означающая 'to be determined'. Она показывает, что значение не определяется из-за малого количества полученных рецензий. Выставлять оценку игре по малой выборке некорректно, поэтому заменим её на NaN. (NaN в этом столбце вероятнее всего означает, что для неё нет вообще ни одной оценки, так что такая замена будет адекватна. 

```python
games_df.loc[games_df['user_score']=='tbd',['user_score']] = np.nan
```

Количество пропусков увеличилось, однако теперь данные можно привести к численному типу и адекватно обработать. 

5. В столбце rating от ESRB так же присутствуют пропуски. Возможно это значит, что рейтинг данным играм не присвоен. Присваивать его самовольно было бы ошибкой. Так или иначе данный столбец нам вообще не особенно интересен, поэтому проигнорируем эти пропуски.

Изменим тип данных в столбец 'year_of_release' на целочисленный.

```python
games_df['year_of_release'] = games_df['year_of_release'].astype('int')
```

А тип данных в столбце 'user_score' на число с плавающей запятой

```python
games_df['user_score'] = games_df['user_score'].astype('float')
```

- **Добавление новых данных**

Посчитаем суммарные продажи по всем регионами для каждой игры.

```python
games_df['total_sales']=games_df[[x for x in games_df.columns if 'sales' in x]].sum(axis=1)
```

По графику хорошо видно смену поколений различных консолей. 

До 1993 года количество выпускаемых игр было невелико. Рост числа релизов в это время практически не наблюдался. 

С 1993 года отсчитывается так называемое пятое поколение игровых систем, включающее в себя Sega Saturn (1994), Sony PlayStation (1994) и Nintendo 64 (1996). Количество выпускаемых игр в это время сильно идет вверх. 

С 2000 года принято выделять шестое поколение консолей. Список платформ 6 поколения включает Sega Dreamcast (1998), Sony PlayStation 2 (2000), Nintendo GameCube (2001) и Microsoft Xbox (2001). Количество релизов по сравнению с 5 поколением увеличивается практически в 2 раза

В 2005 году выходит косоль Microsoft Xbox 360, положившая начало седьмому поколению.  Помимо неё к  консолям 7 поколения относят Sony PlayStation 3 (2006) и, с оговорками, Nintendo Wii (2006). Именно на этот период приходится пик количества выпущенных игр. Консоли были достаточно дешевы, к тому же все ещё продолжался выпуск консолей 6 поколения. 

В 2012 начинаются продажи консолей восьмого (актуального) поколения. Это не очень популярная Nintendo Wii U (2012), а так же Sony PlayStation 4 (2013) и Xbox One (2013). Игр для этого поколения консолей выпущено гораздо меньше чем для предыдущего. Вероятно это связано с тем, что разработка игр становится очень дорогим удовольствием. Как видно, структура рынка при каждом поколении значительно меняется и, если брать данные за все время выводы будут некорректны. Сформируем таблицу с данными игр за актуальный период.

За актуальный период выпущено всего 1698 игр. Это значительно меньше чем в за все время.  

Построим график распределения игр по годам актуального периода.

```python
ax = current_period_df.pivot_table(
    index='year_of_release', columns='platform', values='total_sales', aggfunc='sum').plot();
ax.set_title(label='Продажи за актуальный период', fontsize=24);
ax.set_xlabel(xlabel='Год', fontsize=16);
ax.set_ylabel(ylabel='Продажи игр по годам, млн долларов', fontsize=16);
```

```python
current_platform_pivot = current_period_df.pivot_table(index='platform', columns='year_of_release', 
                                                      values='total_sales', aggfunc='sum').fillna(0)
```

```python
current_platform_pivot
```

Видно, что с выходом нового поколения консолей продажи игр предыдущего поколения стремительно падают. Продажи игр для консолей нового поколения растут (так как данные за 2016 год не полные, наблюдается некоторое снижение общих продаж).

```python
result = current_period_df.groupby(["platform"])['total_sales'].aggregate(
    'count').reset_index().sort_values('total_sales', ascending=False).head(10)
ax = sns.barplot(x='platform', y ='total_sales', 
            data=current_period_df[current_period_df.platform.isin(current_platform_pivot.index.values)], 
            estimator=len, order=result['platform']);
ax.set_title(label='Продажи по платформам за актуальный период', fontsize=24);
ax.set_xlabel(xlabel='Платформа', fontsize=16);
ax.set_ylabel(ylabel='Суммарные продажи', fontsize=16);
```

На удивление неплохо показывают себя портативные консоли PS Vita и 3DS, хотя и продажи игр для них снижаются. 
Потенциально самыми прибыльними платформами, показывающими уверенный рост являются PS4 и Xbox One.

## Шаг 5. Проверка гипотез

- Проверим гипотезу о том, что средние пользовательские рейтинги платформ Xbox One и PC одинаковые.  

Посмотрим на распределения пользовательского рейтинга по основным платформам за актуальный период

```python
sns.boxplot(y="platform", x="user_score", 
            data=current_period_df[current_period_df.platform.isin(top_current_platforms.index.values)], orient="h");
```

Визуально можно заметить, что средний пользовательский рейтинг на платформах PC и Xbox One примерно одинаков. При этом на PC наблюдается больший хвост в сторону игр меньшей оценкой. 

Посчитаем среднюю оценку пользователей по платформам. 

```python
current_period_df[current_period_df.platform.isin(top_current_platforms.index.values)].pivot_table(
    index='platform', values='user_score', aggfunc=('mean', 'median', 'count'))
```

Можно заметить, что медианный рейтинг у обоих платформ одинаков. Это значит, что на среднее в случае с PC влияет большее количество игр с плохим рейтингом. 

Значения пользовательского рейтинга для многих записей в таблице отсутствуют. Это не значит, что их нет совсем, однако означает, что выборка у нас не полная. 

Исходя их этого можно сформировать нулевую гипотезу. Так как как правило нулевая гипотеза должна включать в себя условие равенства, она будет звучать так - средние пользовательские рейтинги платформ Xbox One и PC одинаковы. Соответственно альтернативная гипотеза - средние пользовательские рейтинги платформ Xbox One и PC различны. 

Проверим гипотезу с помощью метода ttest_ind из пакета scipy, проверяющего одинаковость среднего значения двух независимых выборок. Критический уровень статистической значимости зададим равным 5%. Так как в некоторых случаях пользовательские рейтинги отсутствуют, что помешает анализу - уберем пропущенные значения.

```python
alpha = .05
results = st.ttest_ind(current_period_df[current_period_df['platform']=='PC']['user_score'].dropna(), 
                       current_period_df[current_period_df['platform']=='XOne']['user_score'].dropna())
            
print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Нулевая гипотеза не была отвергнута. Действительно, можно предположить, что средние значения пользовательского рейтинга для обоих платформ одинаковы. Однако p-значение мало, и если бы мы задали критический уровень статистической значимости равный 10% гипоезу пришлось бы отвергнуть.

Можем предположить, что дело в том, что и на PC и на Xbox встречаются эксклюзивы. Что если взять игры, которые выходили только на обоих этих платформах? Тогда мы сможем узнать, одинакого ли оценивают одни и те же игры и узнать, влияет ли сама платформа на пользовательский рейтинг мультиплатформенной игры.  

Составим списки с названиями игр, выходивших на PC и XOne

```python
pc_games_list = current_period_df[current_period_df['platform']=='PC']['name'].tolist()
XOne_games_list = current_period_df[current_period_df['platform']=='XOne']['name'].tolist()
```

И сформируетм нулевую гипотезу. сформировать нулевую гипотезу. Она будет звучать так - средние пользовательские рейтинги игр, выходящих одновременно на Xbox One и PC одинаковы. Соответственно альтернативная гипотеза - средние пользовательские рейтинги игры, выходящих одновременно на Xbox One и PC различны.

Проверим гипотезу с помощью метода ttest_ind из пакета scipy

```python
alpha = .05
results = st.ttest_ind(current_period_df[(current_period_df['platform']=='PC') & (current_period_df['name'].isin(XOne_games_list))]['user_score'].dropna(), 
                       current_period_df[(current_period_df['platform']=='XOne') & (current_period_df['name'].isin(pc_games_list))]['user_score'].dropna())
            
print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Получается, что нулевая гипотеза отвергнута, и в среднем рейтинги игры, выходящей одновременно на двух этих платформах все же различаются. 

Посчитаем это количественно

```python
pcxbox_intersection=list(set(pc_games_list) & set(XOne_games_list))
```

```python
current_period_df[(current_period_df.platform.isin(('PC', 'XOne'))) & 
                  (current_period_df.name.isin(pcxbox_intersection))].pivot_table(
    index='platform', values='user_score', aggfunc=('mean', 'median', 'count'))
```

Можно сделать вывод, что если игра выходит одновременно на PC и XOne, то на PC она получает в среднем меньшие оценки.

Мы отвергли нулевую гипотезу. Значит средние пользовательские рейтинги жанров Sport и Action все же различны. При этом мы можем ясно видеть, что средний пользовательский рейтинг у жанра Sport значительно ниже.
