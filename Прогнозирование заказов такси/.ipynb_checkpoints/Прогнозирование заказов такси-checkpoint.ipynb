Наша модель не стационарна и содержит много шума, поэтому использовать численные методы анализа временных рядов вроде ARIMA не представляется возможным. В данной работе мы будем использовать модель градиентного бустинга, для которой предварительно подготовим необходимые фичи, которые будут описывать тренд и сезонность.

Создадим следующие фичи:

1. Календарные признаки  

Это месяц, день, день недели. Возможно в них нет большой необходимости, ведь промежуток времени, который нам необходимо изучить не велик, однако проверить их все равно стоит.

Количество отстающих значений выбирается в зависимости от сезонности. В данном случае очевдиным решением будет взять 24 таких значения, так как на графике сезонности наблюдается именно такой период, то есть значение в необходимый нам час должно хорошо предсказываться значением в этот же час за предыдущий день. ### <u>КОММЕНТАРИЙ РЕВЬЮЕРА</u>

<font color='green'>
Lag: его называют ещё параметром авторегрессии. На бытовом уровне его можно понять так: допустим вчера мы собрались с друзьями и не много закусили.... повлияет ли это на сегодняшний настрой на жизнь? Скорее да)))) Так вот с авторегрессией также. Допустим lag=1 - учитывает (для удобства будем считать, что продажи собраны по дням) как вчерашние продажи - вчерашние события - воалияли на сегоняшние.Lag=2: учитывает как позавчерашние продажи влияютна сегодняшние.....<br>
    
Поэтому, в целом: lag НЕ выбирается от сезонности. За это отвечают факторы сезонности. lag - это влияние вчера на сегодня. Предыдущего значения на последующее.<br>
Хотя, положа руку на сердце, могу сказать следующее: если в модели НЕ учли фактор сезонности, например суточную - 24 часа. То lag =24 сможет эту "оплошность" исправить. Но, т.к. мы учитываем сезонность, то нет смысла брать большой лаг. Ведь и бизнесово это сложно объяснить: лаг=19 - событие 19 часов назад влияет на текущий час. И так каждый час. то это такое может быть?

Скользящее средннее вероятнее всего имеет смысл взять за предыдущую неделю и предыдущие 24 часа, что будет соответствовать сезонным колебаниям.

<font color='blue'> 
### <u>КОММЕНТАРИЙ РЕВЬЮЕРА</u>
</font>
<font color='green'>
    
Например, rolling=10 - скользящее среднее. По сути, ищем среднее по 10 точкам, потом следующие 10 точек, и т.д. - это некое подобие тренда будет - средней линии. Это легко увидеть на пределе: допустим длина ряда 100 дней. rolling(100) - это просто одно число - среднее всего ряда. <br>
    
В общем-то, это можно интерпертировать как тренд. Но в модель можно вставлять тренд из декомпозиции тоже. ТОгда скользящее среднее и не нужно)

___

Напишем функцию для создания необходимых нам признаков.

```python
def make_features(data):
    data['month'] = data.index.month
    data['day'] = data.index.day
    data['dayofweek'] = data.index.dayofweek
    
    data['hour'] = data.index.hour
      
    for lag in range(1, 25):
        data['lag_{}'.format(lag)] = data['num_orders'].shift(lag)
        
    data['lag_week'] = data['num_orders'].shift(168)

    data['rollingday'] = data['num_orders'].shift().rolling(24).mean()
    data['rollingweek'] = data['num_orders'].shift().rolling(168).mean()
    
    return data
```

```python
taxi_time_series = make_features(taxi_time_series)
```

Удалим NaN, которые образовались при вычислении лага и скользящего среднего.

```python
taxi_time_series.dropna(inplace=True)
```

Теперь наши данные готовы к обучению. 

Выделим целивой признак

```python
X = taxi_time_series.drop('num_orders', axis=1)
y = taxi_time_series['num_orders']
```

После выделим тренировочную часть датасета, без перемешивания, так, чтобы в тестовой части оказались 10% последних значений.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, shuffle=False, test_size=0.1, random_state=42)
```

Обучим LightGBM на тестовом датасете.

```python
train_data = lightgbm.Dataset(X_train, label=y_train)
test_data = lightgbm.Dataset(X_test, label=y_test)
```

Зададим параметры обучения

```python
parameters = {
    'objective': 'regression', 
    'metric': 'l2', 
    'boosting': 'gbdt', 
    'learning_rate': 0.1,
    'verbose': 1,
    'stratified': False,
}
```

```python
%%time
model_lgbm = lightgbm.train(parameters,
                       train_data,
                       num_boost_round=200)
```

# 4. Тестирование

```python
def rmse(y_true, y_pred):
    return mean_squared_error(y_true, y_pred) ** 0.5
```

Нам необходимо достичь RMSE не более 48. Посмотрим, для проверки на адекватность, как с этим справляется константная модель, предсказывающая значения по медиане.

```python
pred_median = np.ones(y_test.shape) * np.median(y_train.values)
```

```python
rmse(y_test, pred_median)
```

Недостаточно хорошо. (вероятнее всего это можно объяснить достаточно резким трендом к концу данных. Это хитро. Мне кажется предсказание по медиане сработало бы лучше если бы в данных не появлялся резкий рост и выбросы как раз в тестовой части. 

Проверим так же, насколько хорошо работает предсказание по предыдущему значению.

```python
pred_previous = y_test.shift()
pred_previous.iloc[0] = y_train.iloc[-1]
```

```python
rmse(y_test, pred_previous)
```

Гораздо лучше! Хотя до цели не хватает. 

Проверим, наконец, как работает модель с градиентным бустингом. 

```python
y_lgbm_pred = model_lgbm.predict(X_test)
```

```python
rmse(y_test, y_lgbm_pred)
```
